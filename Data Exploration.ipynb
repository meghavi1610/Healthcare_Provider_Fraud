```python
import pandas as pd
import networkx as nx

# Assuming df is your DataFrame with columns 'Provider', 'AttendingPhysician', and 'Beneficiary'
# Create a graph
G_last = nx.from_pandas_edgelist(df, source='Provider', target='BeneID')

# Calculate network features
provider_degree = dict(G_last.degree())
ap_degree = nx.degree_centrality(G_last)
ap_closeness = nx.closeness_centrality(G_last)
ap_eigenvector = nx.eigenvector_centrality(G_last ,max_iter=1000, tol=1e-6)

# Create a new DataFrame to store network features
features_df = pd.DataFrame({
    'Provider': list(provider_degree.keys()),
    'ProviderDegree': list(provider_degree.values()),
    'APDegreeCentrality': [ap_degree[node] for node in provider_degree.keys()],
    'APClosenessCentrality': [ap_closeness[node] for node in provider_degree.keys()],
    'APEigenvectorCentrality': [ap_eigenvector[node] for node in provider_degree.keys()],

})
```

```python
# Filter the DataFrame for source nodes (providers)
source_nodes = df['Provider'].unique()
source_features_df = features_df[features_df['Provider'].isin(source_nodes)]
```
